# Аппроксимация функции

```python
import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
def chlen(x, w):
    result = 0
    for i in range(len(w)):
        result += w[i] * (x ** i)
    return result
```

```python
A = [[1, 1], [1, 15]]
b = np.array([f(1), f(15)])

result1 = linalg.solve(A, b)

space = np.linspace(1, 15, 100)
y1 = []
for i in space:
    y1.append(f(i))
y2 = []
for i in space:
    y2.append(chlen(i, result1))
plt.plot(space, y1)
plt.plot(space, y2)
plt.grid()
plt.show()
```

```python
A = [[1, 1, 1], [1, 8, 8 ** 2], [1, 15, 15 ** 2]]
b = np.array([f(1), f(8), f(15)])

result2 = linalg.solve(A, b)

y2 = []
for i in space:
    y2.append(chlen(i, result2))
plt.plot(space, y1)
plt.plot(space, y2)
plt.grid()
plt.show()
```

```python
A = [[1, 1, 1, 1], [1, 4, 4 ** 2, 4 ** 3], [1, 10, 10 ** 2, 10 ** 3], [1, 15, 15 ** 2, 15 ** 3]]
b = np.array([f(1), f(4), f(10), f(15)])

result3 = linalg.solve(A, b)

y2 = []
for i in space:
    y2.append(chlen(i, result3))
plt.plot(space, y1)
plt.plot(space, y2)
plt.grid()
plt.show()
```
